```python
# 2. Bar Chart: Total Sales dan Profit per Wilayah

plt.figure(figsize=(8, 5))
df_grouped = df.groupby("Region")[["Sales", "Profit"]].sum().reset_index()
df_grouped.plot(x='Region', kind="bar", stacked=False, colormap="coolwarm", figsize=(8, 5))

plt.title("Total Sales dan Profit per Wilayah")
plt.xticks(rotation=0)
plt.xlabel("Region")
plt.ylabel("Total Value")
plt.legend(["Sales", "Profit"])

for i, region in enumerate(df_grouped["Region"]):
    total_values = df_grouped['Sales'][i] + df_grouped['Profit'][i]
    for j, col in enumerate(["Sales", "Profit"]):
        value = df_grouped.loc[i, col]
        percent = (value / total_values) * 100
        plt.text(i + j * 0.2 - 0.1, value + total_values * 0.01, f"{percent:.1f}%", ha='center', fontsize=10, color='black')
plt.show()
```

Insight:

- Wilayah dengan Penjualan & Profit Tertinggi
    + EMEA memiliki total sales tertinggi dan profit tertinggi.

```python
# 3. Perbandingan Rata-rata Sales dan Profit pada Setiap Tingkat Diskon 
df_finance = df[df['Industry']=='Finance']
df_grouped = df_finance.groupby("Discount")[["Profit", "Sales"]].mean().reset_index()

plt.figure(figsize=(8,5))
df_grouped.plot(x="Discount", kind="bar", stacked=False, figsize=(10,5),)
plt.title("Rata-rata Profit dan Sales per Diskon")
plt.xlabel("Discount (%)")
plt.ylabel("Rata-rata Nilai")
plt.legend(["Profit", "Sales"])
plt.show()
```

Insight: Pengaruh Diskon terhadap Sales dan Profit
- Diskon sekitar 10% memberikan sales tertinggi dengan profit masih positif, menunjukkan titik optimal.
- Mulai diskon di atas 30%, profit mulai negatif, meskipun sales tetap tinggi. Artinya, diskon besar memang menarik pembeli, tapi merugikan perusahaan.
- Diskon terlalu tinggi (>50%) menyebabkan kerugian besar tanpa peningkatan signifikan pada sales.

Kesimpulan: Diskon kecil-menengah efektif menaikkan penjualan tanpa merusak profit. Hindari diskon besar karena justru bisa bikin rugi.

## **Hypotesis**


Berikut analisis mengenai kelayakan untuk pengujian hipotesis:

1. Rata-rata Profit Margin antar Kelompok Produk
    - Hipotesis Nol (H₀): Tidak ada perbedaan signifikan dalam rata-rata profit margin antara produk yang menguntungkan dan produk yang merugi.
    - Hipotesis Alternatif (H₁): Ada perbedaan signifikan dalam rata-rata profit margin antara produk yang menguntungkan dan produk yang merugi.

2. Total Sales dan Profit per Wilayah
    - Membandingkan total sales dan profit berdasarkan wilayah.
    - Jika ada hipotesis seperti "Apakah terdapat perbedaan signifikan antara sales dan profit di berbagai wilayah?", maka cocok untuk diuji dengan ANOVA (lebih dari 2 Variable).

3. Rata-rata Profit Margin dan Jumlah Penjualan per Diskon
    - Menunjukkan hubungan antara diskon dengan profit margin dan jumlah penjualan.
    - Jika ada hipotesis seperti "Apakah ada korelasi signifikan antara diskon dan profit margin?", bisa diuji dengan regresi atau korelasi Spearman.

```python
# Rata-rata Profit Margin antar Kelompok Produk

from scipy.stats import ttest_ind

profit = df[df["Profit Margin %"] > 0]["Profit Margin %"]
loss = df[df["Profit Margin %"] <= 0]["Profit Margin %"]

# Independent t-test
stat, p_value = ttest_ind(profit, loss, equal_var=False)

print(f"stats: {stat}, P-value: {p_value}")

# Interpretasi
if p_value < 0.05:
    print("Menolak H0, ada perbedaan signifikan antara produk yang menguntungkan dan merugi.")
else:
    print("Gagal Menolak H0, tidak ada perbedaan signifikan antara produk yang menguntungkan dan merugi.")
```

```python
# Total Sales Per Wilayah

from scipy.stats import f_oneway

# Uji ANOVA untuk Sales
stats, pvalue_sales = f_oneway(
    df[df['Region'] == 'AMER']['Sales'],
    df[df['Region'] == 'APJ']['Sales'],
    df[df['Region'] == 'EMEA']['Sales']
)
print(f'Anova pvalue Sales: {pvalue_sales}')

# Uji ANOVA untuk Profit
stats, pvalue_profit = f_oneway(
    df[df['Region'] == 'AMER']['Profit'],
    df[df['Region'] == 'APJ']['Profit'],
    df[df['Region'] == 'EMEA']['Profit']
)
print(f'Anova pvalue Profit: {pvalue_profit}')
```

```python
alpha = 0.05
if pvalue_sales < alpha:
    print(" Hasil Sales: Menolak H0, varians kedua kelompok berbeda.")
else:
    print(" Hasil Sales: Gagal menolak H0, varians kedua kelompok dianggap sama.")

alpha = 0.05
if pvalue_profit < alpha:
    print("Hasil Profit: Menolak H0, varians kedua kelompok berbeda.")
else:
    print("Hasil Profit: Gagal menolak H0, varians kedua kelompok dianggap sama.")
```

```python
# Rata-rata profit margin per Quantity

from scipy.stats import shapiro

stat, pvalue = shapiro(df['Profit Margin %'])
print(f"Shapiro-Wilk Test: Statistik={stat}, p-value={pvalue}")
```

```python
# Distribusi Profit Margin Berdasarkan Quantity
from scipy.stats import spearmanr

# Uji Korelasi Spearman (Data tidak berdistribusi normal)
spearman_corr, spearman_p = spearmanr(df['Quantity'], df['Profit Margin %'])
print(f"Spearman Correlation: {spearman_corr}, p-value: {spearman_p}")

```

```python
# Interpretasi hasil
alpha = 0.05  # Tingkat signifikansi 5%

if spearman_p < alpha:
    print("Terdapat hubungan signifikan antara Quantity dan Profit Margin (Spearman)")
else:
    print("Tidak ada hubungan signifikan antara Quantity dan Profit Margin (Spearman)")
```
